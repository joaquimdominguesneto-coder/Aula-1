# 🚀 Python Avançado - Visualizações e Exemplos

Notebook com gráficos, análise de dados e exemplos mais avançados.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## 📊 Gráfico Simples

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure()
plt.plot(x, y)
plt.title("Gráfico de Linha")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## 📊 Gráfico de Barras

```python
categorias = ["A", "B", "C"]
valores = [10, 20, 15]

plt.figure()
plt.bar(categorias, valores)
plt.title("Gráfico de Barras")
plt.show()
```

## 📊 Análise com Pandas

```python
dados = {
    "nome": ["Ana", "João", "Maria"],
    "idade": [20, 25, 30]
}

df = pd.DataFrame(dados)
print(df)
print("Média de idade:", df["idade"].mean())
```

## 📊 Histograma

```python
idades = [20, 21, 22, 23, 24, 25, 30, 35]

plt.figure()
plt.hist(idades)
plt.title("Distribuição de Idades")
plt.show()
```

## 🔬 Simulação

```python
import random

resultados = [random.randint(1, 6) for _ in range(100)]

plt.figure()
plt.hist(resultados)
plt.title("Simulação de Dados (Dado)")
plt.show()
```

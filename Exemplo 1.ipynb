# 🐍 Python - Exemplos Práticos

Este notebook apresenta exemplos práticos de aplicações em Python.

## 1. Básico

```
print("Olá, mundo!")
```

## 2. Variáveis

```
nome = "Ana"
idade = 20
print(nome, idade)
```

## 3. Entrada de Dados

```
nome = "Usuário"  # simulação de input
print(f"Olá, {nome}!")
```

## 4. Condicionais

```
idade = 18
if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

## 5. Laços de Repetição

```
for i in range(5):
    print(i)
```

## 6. Funções

```
def soma(a, b):
    return a + b

print(soma(2, 3))
```

## 7. Listas

```
numeros = [1, 2, 3, 4]
print(sum(numeros))
```

## 8. Dicionários

```
pessoa = {"nome": "Ana", "idade": 20}
print(pessoa["nome"])
```

## 9. Mini Projeto

```
notas = [7, 8, 9]
media = sum(notas) / len(notas)
print("Média:", media)
```
